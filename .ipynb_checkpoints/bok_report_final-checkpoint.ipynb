```python
from selenium import webdriver
import time
from tika import parser
import pandas as pd
import re
driver = webdriver.Chrome('C:\chromedriver.exe')
```

```python
import numpy as np
page = np.arange(1,107)
link_ls = []
date_ls = []
title_ls = []
```

```python
for p in page:
    url = 'https://finance.naver.com/research/debenture_list.nhn?keyword=&brokerCode=&searchType=writeDate&writeFromDate=2010-01-01&writeToDate=2020-08-31&page='+str(p)
    driver.get(url)
    links = driver.find_elements_by_xpath('//*[@id="contentarea_left"]/div[3]/table[1]/tbody/tr/td[3]/a')
    dates = driver.find_elements_by_xpath('//*[@id="contentarea_left"]/div[3]/table[1]/tbody/tr/td[4]')
    titles = driver.find_elements_by_xpath('//*[@id="contentarea_left"]/div[3]/table[1]/tbody/tr/td[1]/a')
    
    for n in range(len(titles)):
        title_ls.append(titles[n].text)
        date_ls.append(dates[n].text)
        link_ls.append(links[n].get_attribute("href"))
    print(str(p)+' page 끝')
driver.close()
```

```python
len(date_ls)
```

```python
# pdf 링크 저장
with open('./tika_report/links.txt', 'w', encoding='utf-8') as f:
    for l in link_ls:
        f.write(l)
with open('./tika_report/dates.txt', 'w', encoding='utf-8') as f:
    for d in date_ls:
        f.write(d)
with open('./tika_report/titles.txt', 'w', encoding='utf-8') as f:
    for t in title_ls:
        f.write(t)
```

```python
# 저장해놓은 링크 불러오기
links = []
with open('./tika_report/links.txt','r') as f:
    strings = f.readlines()
    strings = strings[0].split('pdf')
    links.append(strings)
    f.close()
links = links[0]
links
```

```python
i = 1
final = []
for link,date,title in zip(link_ls,date_ls,title_ls):
    PDFfileName = link
    inputpath = PDFfileName
    parsed = parser.from_file(PDFfileName)
    content = re.sub("[\\n]","",str(parsed['content']))
    df = pd.DataFrame.from_dict([{'date':date,'title':title,'content':content}])
    new_title = re.sub("[\/:*?\"<>|]"," ",title)
    df.to_csv('./tika_report/'+date+' '+new_title+'.csv', 'w', encoding='utf-8')
    final.append((date,title,content))
    print(i)
    i += 1
```

```python
final_df = pd.DataFrame(final,columns = ['date','title','content'])
final_df.to_csv('./tika_report/bond_final.csv')
```
