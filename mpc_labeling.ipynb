```python
import pandas as pd
```

```python
clean = pd.read_csv('mpc_ngram.csv', encoding='utf-8')
```

```python
# 제가 가지고 있던 파일로 진행해서 'date' 바꿔준 것이니 실제로는 사용 안 하셔도 됩니다

clean['newdate'] = clean['date'].str.replace('.','-')
clean = clean.drop(['date'], axis=1)
```

```python
call_rate = pd.read_csv('labeled_callrate.csv')
```

```python
final = pd.merge(clean,call_rate[['DATE','label']],left_on='newdate', right_on='DATE')
```

```python
# NAN 값 있는 애들 없애주기
final = final.fillna('')
final = final[final['ngrams'] != '']
```

```python
final.to_csv('mpc_ngram_label.csv', encoding='utf-8')
```
